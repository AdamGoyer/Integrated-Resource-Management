# Client Onboarding Template

### Financial Planning Client Onboarding Framework

#### **Objective:**
To create an engaging and efficient client onboarding experience that combines digital tools and conversational AI to collect comprehensive financial information while ensuring a client-centric approach.

---

### **Scope of the Project**
1. **Objective:** To develop a digital and conversational onboarding system that:
   - Collects detailed client financial and personal data.
   - Simulates a human advisor’s conversational style.
   - Generates a client file for analysis and planning.
   - Seamlessly integrates with existing financial tools (e.g., eMoney, MassMutual).

2. **Technological Approach:**
   - Use a combination of real-time AI tools (e.g., ChatGPT, Hume.ai) for dynamic conversations.
   - Implement a digital fact-finder as a backup or supplementary tool.
   - Ensure compatibility with mobile, desktop, and tablet devices.

3. **Target Audience:**
   - New clients seeking financial planning.
   - Returning clients updating their information.

4. **Compliance:**
   - Ensure the system meets regulatory requirements (FINRA, SEC).
   - Incorporate data privacy and security protocols (e.g., GDPR, CCPA).

---

### **Questions and Data Collection Goals**

#### **Personal Information**
1. **Client Details:**
   - Full Name, DOB, Citizenship Status.
   - Address, Contact Information.
   - Employment Information (Employer, Title, Years at Job).

2. **Spouse/Partner Details:**
   - Full Name, DOB, Citizenship Status.
   - Employment Information.

3. **Family Details:**
   - Children (Names, DOBs, Dependency Status).
   - Other Dependents or Special Circumstances.

#### **Financial Overview**
1. **Income:**
   - Annual Salary, Bonuses, Other Income.
   - Social Security Benefits (Current or Projected).
   - Pension Details.

2. **Expenses:**
   - Monthly Living Expenses.
   - Discretionary Spending (e.g., vacations, hobbies).
   - Anticipated Changes in Retirement.

3. **Assets:**
   - Real Estate (Address, Current Value, Purchase Price).
   - Investments (401(k), IRAs, Brokerage Accounts).
   - Cash Accounts (Checking, Savings).

4. **Liabilities:**
   - Mortgages (Balance, Term, Interest Rate).
   - Other Loans (Student, Auto, HELOC).
   - Credit Card Balances.

#### **Insurance**
1. **Current Policies:**
   - Life Insurance (Term, Whole, Death Benefit).
   - Disability Insurance (Benefit Amount, Waiting Period).
   - Long-Term Care (Coverage Amount, Premiums).

2. **Future Needs:**
   - Is current coverage sufficient?
   - Interest in reviewing additional options?

#### **Goals and Risk Tolerance**
1. **Short-Term Goals:**
   - Debt Reduction, Emergency Fund.
   - Major Purchases (Home, Vehicle).

2. **Long-Term Goals:**
   - Retirement Age and Lifestyle Expectations.
   - Education Funding for Children.
   - Estate Planning and Legacy Goals.

3. **Risk Tolerance:**
   - Conservative, Moderate, Aggressive.

#### **Estate Planning**
1. **Current Documents:**
   - Wills, Trusts, Power of Attorney.

2. **Future Planning:**
   - Gifting or Charitable Contributions.
   - Guardianship Provisions.

#### **Advisor-Specific Questions**
1. **Decision-Making:**
   - Do you work with a CPA or attorney?
   - Are these professionals key decision-makers?

2. **Obstacles:**
   - What’s your biggest challenge in achieving your financial goals?
   - Are you willing to invest effort or funds in tax-saving strategies?

---

### **Conversational AI Features**
1. **Dynamic Questions:**
   - Adjust follow-up questions based on client responses.
   - E.g., If a client has no dependents, skip related sections.

2. **Empathy and Engagement:**
   - Use conversational tones to put clients at ease.
   - Provide positive feedback (e.g., “That’s a great start!”).

3. **Real-Time Clarifications:**
   - Clarify ambiguous answers immediately (e.g., “When you say you’re planning to retire early, what age are you thinking of?”).

4. **Summary and Confirmation:**
   - Summarize collected data at the end for client approval.
